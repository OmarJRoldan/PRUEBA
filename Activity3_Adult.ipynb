```python
df.rename(columns = {'marital-status':'marital_status'})
```

```python
mstatus = pd.pivot_table(df,
               index = 'marital-status',
               aggfunc = 'count', 
               values = 'education'
               )
mstatus
```

```python
mstatus['% of Total'] = (mstatus.education / mstatus.education.sum() * 100)
print (mstatus)
```

```python
pd.pivot_table(df,
               index = 'marital-status', 
               values = 'age',
               columns = 'education'
               )
```

```python
df.groupby(['marital-status', 'education'])[['hours-per-week']].mean()
```
